## Machine Status Dummy

```python
data['machine_status'].value_counts()
```

```python
mac_stat_uniq = list(data['machine_status'].value_counts().index)
```

```python
mac_stat_uniq
```

```python
def mac_status_dummy(col):
    return mac_stat_uniq.index(col)
```

```python
data['machine_status'] = data['machine_status'].apply(mac_status_dummy)
```

```python
data['machine_status'].value_counts()
```

```python
data.drop(['timestamp'],axis=1,inplace=True)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['machine_status'],axis=1)
```

```python
y = data['machine_status']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=100,stratify=y)
```

```python
X_train.shape,y_train.shape
```

# Machine Learning

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
pred1 = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred1))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
pred2 = dt.predict(X_test)
```

```python
print(classification_report(y_test,pred2))
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
print(classification_report(y_test,pred3))
```
