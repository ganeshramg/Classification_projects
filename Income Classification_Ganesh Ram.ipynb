## Dummies(Converting strings and categories to intergers)

```python
data.head()
```

```python
data.columns
    
```

```python
num_cols = data._get_numeric_data()
```

```python
cat_cols = list(set(data.columns)-set(num_cols))
```

```python
cat_cols
```

```python
categories = []

def get_dummy(col):
    return categories.index(col)

for column in cat_cols:
    categories = list(data[column].value_counts().index)
    data[column] = data[column].apply(get_dummy)
    
# For every column_name in categorical columns in the dataFrame, 
# Get unique values and save it as a list of categories
# Then apply the function to use the index value of that category itself as a dummy
```

```python
data.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['income'],axis=1)
```

```python
y = data['income']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y)
```

```python
X_train.shape,y_train.shape
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
error = []

for i in range(1,200):
    rf_i = RandomForestClassifier(n_estimators=i)
    rf_i.fit(X_train,y_train)
    pred_i =rf_i.predict(X_test)
    error.append(np.mean(y_test!=pred_i))
    
plt.plot(error)
```

```python
print(error.index(min(error)))
```

```python
rf = RandomForestClassifier(n_estimators=67)
```

```python
rf.fit(X_train,y_train)
```

```python
pred1 = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred1))
```
